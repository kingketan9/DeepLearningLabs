```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Ques 1:


```python
mean = 0
std_dev = 1

expectations = []
for N in range(100, 10 ** 5, 1000):
    random_no = np.random.normal(loc = mean, scale = std_dev, size = (N,))
    
    expectation = np.sum(np.logical_and(random_no <= 1, random_no >= -1)) / N
    expectations.append(expectation)
      
_, bins, _ = plt.hist(random_no, 100, density = True)
plt.plot(bins, norm.pdf(bins, loc = mean, scale = std_dev), linewidth = 2, color='#DC143C')
plt.show()
```

```python
E_X = 68.2
MEAN_E_X = np.mean(expectations) * 100


MEAN_E_X = int(MEAN_E_X * 10) / 10

if E_X == MEAN_E_X:
    print('Result Verified.')
    print(f'mean(XN) = E(X) = {E_X}%')
else:
    print('Your results are wrong. Check again!')
    print(f'mean(XN) != E(X) = {E_X}%')
```

```python
# 2(i)
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
month_dict = {0:'Jan',1:'Feb',2:'March',3:'April',4:'May',5:'June',6:'July',7:'August',8:'Sept',9:'Oct',10:'Nov',11:'Dec'}
revenue = np.array(revenue)
expenses = np.array(expenses)
profit = revenue-expenses  
np.round(profit, 2) 
```

```python
# 2(ii) Profit after tax for each month (the tax rate is 30% to revenue)
tax = 0.3*revenue
profit_after_tax = profit-tax
np.round(profit_after_tax, 2)
```

```python
#2(iii)
profit_margin = profit_after_tax/revenue*100
np.round(profit_margin, 0) 
```

```python
#2(iv)
print("Good Months ->") 
mean_of_year = profit_after_tax.mean()
months = profit_after_tax > mean_of_year
good_months = []
for i in range(12):
  if months[i]==True:
    good_months.append(month_dict[i])
print(good_months)
```

```python
#2(v)
print("Bad Months ->") 
bad_months=[]
for i in range(12):
  if months[i]==False:
    bad_months.append(month_dict[i])
print(bad_months)
```

```python
#2(vi)
worst_month = profit_after_tax==profit_after_tax.min() 
print("Worst Month ->")
for i in range(12):
  if worst_month[i]==True:
    print(month_dict[i])
```

```python
#2(vii)
best_month = profit_after_tax==profit_after_tax.max() 
print("Best Month ->")
for i in range(12):
  if best_month[i]==True:
    print(month_dict[i])
```
