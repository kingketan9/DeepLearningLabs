```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Set the number of input and output channels, kernel size, stride, and padding
inChans = 3
outChans = 15
knrSize = 5
stride = 1
padding = 0

# Create a convolutional layer using the nn.Conv2d class
conv_layer = nn.Conv2d(inChans, outChans, knrSize, stride, padding)

# Print the convolutional layer
print(conv_layer)
print(' ')

# Print the size of the weights and bias of the convolutional layer
print("Size of weights: " + str(conv_layer.weight.shape))
print("Size of bias: " + str(conv_layer.bias.shape))

# Plot the weights of the convolutional layer
fig, axs = plt.subplots(3, 5, figsize=(10, 5))

for i, ax in enumerate(axs.flatten()):
    ax.imshow(torch.squeeze(conv_layer.weight[i, 0, :, :]).detach(), cmap='Purples')
    ax.set_title('L1(0) -> L2(%s)' % i)
    ax.axis('off')

plt.tight_layout()
plt.show()

# Create an input image
img = torch.randn(1, inChans, 64, 64)

# Show the input image
img2view = img.permute(2, 3, 1, 0).numpy()
plt.imshow(np.squeeze(img2view))

# Pass the input image through the convolutional layer
conv_res = conv_layer(img)

# Print the size of the input and output tensors
print("Input shape: ", img.shape)
print("Output shape: ", conv_res.shape)

# Plot the output of the convolutional layer
fig, axs = plt.subplots(3, 5, figsize=(10, 5))

for i, ax in enumerate(axs.flatten()):
    I = torch.squeeze(conv_res[0, i, :, :]).detach()
    ax.imshow(I, cmap='Purples')
    ax.set_title('Conv. w/ filter %s' % i)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
